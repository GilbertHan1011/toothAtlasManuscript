```python
meta = meta.drop(columns=meta.filter(like='Unnamed').columns)
meta.columns
```

```python
# drop the index with na
meta = meta[~meta.index.isna()]
```

```python
for cat in meta.columns:
    if meta[cat].isnull().any():
        print(cat, "has null values")
    elif "nan" in meta[cat]:
        print(cat, "has 'nan' values")
```

```python
## Temporally replace na value
meta = meta.fillna("Not provided")
```

```python
for cat in meta.columns:
    if meta[cat].isnull().any():
        print(cat, "has null values")
    elif "nan" in meta[cat]:
        print(cat, "has 'nan' values")
```

```python
meta[0:5]
```

```python
meta["Sample"] = meta.index
```

```python
scMeta = scMeta.reset_index()
```

```python
metaUpdate = pd.merge(scMeta, meta, on="Sample",how = "left")
```

```python
metaUpdate[0:4]
```

```python
metaUpdate.set_index('index', inplace=True)
```

```python
metaUpdate.to_csv("../../processed_data/metadata/scMetadata_latest.csv")
```
